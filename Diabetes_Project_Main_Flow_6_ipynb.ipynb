```python
#Let's find the answer of first question.

#"1.How many people have diabetes and how many do not?"

# Count the number of people with and without diabetes
diabetes_data_counts = data['Diagnosis'].value_counts()
```

```python
#gatting the values
data.Diagnosis.value_counts()
```

```python
#  Bar chart Plotting
plt.figure(figsize=(8, 6))

sns.barplot(x=diabetes_data_counts.index, y=diabetes_data_counts.values, hue=diabetes_data_counts.index, palette='Blues', legend=False)


#Adding labels and title
plt.xlabel('Diabetes Outcome', fontsize=12)
plt.ylabel('Number of people', fontsize=12)
plt.title('Number of people with and without Diabetes',fontsize=16)
plt.xticks([0, 1], ['No Diabetes', 'Diabetes'])
plt.show()
```

```python
# Labels for the pie chart
labels = ['No Diabetes', 'Diabetes']

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(diabetes_data_counts, labels=labels, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'orange'])

# Adding a title
plt.title('Distribution of People with and without Diabetes', fontsize=16)
plt.show()
```

```python

# Plotting the count plot
plt.figure(figsize=(8, 6))
sns.countplot(x='Diagnosis', data=data, palette='Set2')

# Adding labels and title
plt.xlabel('Diabetes Outcome', fontsize=12)
plt.ylabel('Number of People', fontsize=12)
plt.title('Count of People with and without Diabetes', fontsize=16)
plt.xticks([0, 1], ['No Diabetes', 'Diabetes'])
plt.show()
```

```python
#Let's find the answer of 2nd question.
#"2.Which age group has the highest prevalence of diabetes?"

# Sample DataFrame
data = pd.DataFrame({
    'AgeGroup': ['20-30', '30-40', '20-30', '30-40', '40-50', '40-50', '50-60', '60-70', '60-70', '70-80'],
    'DiabetesStatus': [1, 0, 1, 0, 0, 1, 1, 1, 0, 0]
})

# Calculate the total count and the count of diabetes cases by age group
age_group_summary = data.groupby('AgeGroup').agg(
    total_count=('DiabetesStatus', 'size'),
    diabetes_count=('DiabetesStatus', 'sum')
)

# Compute the prevalence of diabetes
age_group_summary['Prevalence'] = age_group_summary['diabetes_count'] / age_group_summary['total_count']

# Plotting Bar Chart
age_group_summary['Prevalence'].plot(kind='bar', color='salmon')


plt.title('Prevalence of Diabetes by Age Group', fontsize=12)
plt.xlabel('Age Group')
plt.ylabel('Prevalence of Diabetes')
plt.xticks(rotation=45)  # Rotate labels for better readability
plt.tight_layout()  # Adjust layout to make room for labels
plt.show()

# Identify the age group with the highest prevalence
max_prevalence_age_group = age_group_summary['Prevalence'].idxmax()
print(f"The age group with the highest prevalence of diabetes is: {max_prevalence_age_group}")
```

```python
#Let's find the answer of 4th question.
#"4. Is there a relationship between BMI and diabetes prevalence?"


#  Create BMI categories
bmi_bins = [0, 18.5, 24.9, 29.9, 34.9, 39.9, np.inf]  # BMI categories based on WHO classification
bmi_labels = ['Underweight', 'Normal weight', 'Overweight', 'Obesity I', 'Obesity II', 'Obesity III']
data['BMI_Category'] = pd.cut(data['BMI'], bins=bmi_bins, labels=bmi_labels)

# Calculate diabetes prevalence in each BMI category
bmi_prevalence = data.groupby('BMI_Category')['Diagnosis'].mean()

#  Plot the relationship with different colors
colors = ['lightblue', 'lightgreen', 'gold', 'orange', 'red', 'purple']  # Define a color for each category

plt.figure(figsize=(10, 6))
bmi_prevalence.plot(kind='bar', color=colors)
plt.title('Diabetes Prevalence by BMI Category', fontsize=14)
plt.xlabel('BMI Category', fontsize=12)
plt.ylabel('Diabetes Prevalence (Proportion)', fontsize=12)
plt.xticks(rotation=45)
plt.show()

#  Print out the BMI category with the highest prevalence
max_prevalence_bmi = bmi_prevalence.idxmax()
print(f"The BMI category with the highest diabetes prevalence is: {max_prevalence_bmi}")
```
